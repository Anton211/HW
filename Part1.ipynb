```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
```

```python
Преобразование данных
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('./weatherAUS.csv', sep=',')
df.loc[df['RainTomorrow'] == 'No', 'RainTomorrow'] = 1
df.loc[df['RainTomorrow'] == 'Yes', 'RainTomorrow'] = 2
df.loc[df['RainToday'] == 'No', 'RainToday'] = 1
df.loc[df['RainToday'] == 'Yes', 'RainToday'] = 2
del df['Date']
del df['Evaporation']
del df['Sunshine']
del df['WindGustDir']
del df['WindDir9am']
del df['WindDir3pm']
del df['Cloud9am']
del df['Cloud3pm']
del df['Location']
del df['Rainfall']
del df['RainToday']
df = df.fillna(df.mean())
y = df['RainTomorrow']
del df['RainTomorrow']
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.25, random_state=23, shuffle=False)
X_train = X_train.astype('int')
y_train = y_train.astype('int')
```

```python
Классификация Гаусса
```

```python
%%time
classifier = GaussianNB()
classifier.fit(X_train, y_train)
prediction = classifier.predict(X_test)
print('Gaussian:', np.mean(prediction == y_test))
```

```python
Классификация Бернулли
```

```python
%%time
classifier = BernoulliNB()
classifier.fit(X_train, y_train)
prediction = classifier.predict(X_test)
print('Bernoulli: ', np.mean(prediction == y_test))
```

```python
Логистическая регрессия
```

```python
%%time
classifier = LogisticRegression(penalty='l2', max_iter=300)
classifier.fit(X_train, y_train)
prediction = classifier.predict(X_test)
print('LogRegr: ', np.mean(prediction == y_test))
```

```python
Ближайшие соседи
```

```python
%%time
neigh = KNeighborsClassifier(n_neighbors=100)
neigh.fit(X_train, y_train)
prediction = neigh.predict(X_test)
print('KNeighbors: ', np.mean(prediction == y_test))
```

```python
В процессе выполнения задания была проанализирована и подготовлена к работе таблица данных о погоде.
Построено 4 модели: 
        1)Байесовские классификаторы(классификация Гаусса и Бернулли)
        2)Логистическая регрессия
        3)Метод ближайших соседей
Прогонка методов дала следующие результаты:
        1)Классификация Гаусса выдает точность прогноза ~82.7%
        2)Классификация Бернулли выдает точность прогноза ~76.9%
        По времени выполнения (1) и (2) методы сопоставимы, примерное время 0.1 секунда
        3)Логистическая регрессия имеет переменный параметр max_iter при построении графика зависимости точности прогноза от max_iter
        приходим к выводу, что точность после max_iter=300 составляет ~82.7%
        При увеличении max_iter растет время выполнения, при max_iter=1000 время выполнения составило 2.8 секунд
        4)Метод ближайших соседей выполняется дольше выше разобранных методов и дает результат сопоставимый с (1) и (3)
        При n_neighbors=100 время работы составляет 38 секунд, точность предсказания ~82.9%
Самым оптимальным методом оказалась классификация Гаусса. Он выдет достаточно точное предсказание, не затрачивая большого времени.
```
